#### Train/Test & Normalization

```python
x = df.drop(['Potability'], axis=1)
y = df['Potability']
```

```python
scaler = StandardScaler()
scaler.fit(x);
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=41, shuffle=True, stratify=y)

# Сохраним тестовую выборку на диск:

x_test.to_csv("x_test.csv", index=None)
y_test.to_csv("y_test.csv", index=None)
```

#### Feature engineering

```python
# Воспользуемся RFECV для отбора лучших признаков:

rfecv = RFECV(estimator=RandomForestClassifier())
rfecv.fit(x_train, y_train);
```

```python
rfecv.transform(x_train)
features = list(x.columns[np.where(rfecv.support_ == True)])
features
```

```python
class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key

    def fit(self, x, y=None):
        return self

    def transform(self, x):
        return x[self.key]
```

```python
features_pipeline = Pipeline([
    ('selector', ColumnSelector(features)),
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')),
    ('scaler', StandardScaler())
])

feats = FeatureUnion([
    ('features', features_pipeline)
])
```

#### Pipeline

```python
# Добавляем классификатор (создаём Pipeline):

pipeline = Pipeline([
    ('features', feats),
    ('classifier', RandomForestClassifier()),
])

pipeline.fit(x_train, y_train)
```

#### Сохраняем модель (Pipeline)

```python
with open('app/models/rf_pipeline.dill', 'wb') as f:
    dill.dump(pipeline, f)
```
